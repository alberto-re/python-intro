```python
import requests
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO

%pylab inline
```

```python
def display_img(content):
    img = Image.open(BytesIO(content))
    plt.imshow(img)
    plt.show()
```

```python
response = requests.get("https://api.github.com/repos/tensorflow/tensorflow/events")
response.raise_for_status()

data = response.json()
for entry in data:
    response = requests.get(entry["actor"]["avatar_url"])
    response.raise_for_status()
    display_img(response.content)
```
